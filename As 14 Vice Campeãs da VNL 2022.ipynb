### AVISO: Ao digitar a altura use o ponto(.) ao invés de vírgula (,)

```python
jogadoras = []

for convocadas in range (14):
    nome = input("\nNome da jogadora: ").title().strip()
    numero = int(input('Nº da jogadora: '))
    idade = int(input('Idade : '))
    altura = float(input ('Altura: '))
    posicao = input ('Posição: ').title().strip()
    convocacao = [nome, numero, idade, altura, posicao]
    jogadoras.append(convocacao)   
```

```python
media_idade = 0
media_altura = 0

for nome, numero, idade, altura, posicao in jogadoras:
    media_idade += idade/14
    media_altura += altura/14
    
print('\n')
print ('='*21, "CONVOCADAS", '='*21)
print ('| Nº |      NOME     | IDADE | ALTURA |   POSIÇÃO    |')
print ("-"*54)
for i in jogadoras:
    indice = jogadoras.index(i)
    texto = "|{:^4}|{:<15}|{:^7}|{:^8,.2f}|{:>14}|"
    print (texto.format(jogadoras[indice][1], jogadoras[indice][0], jogadoras[indice][2], jogadoras[indice][3],jogadoras[indice][4]))
print ("-"*54)
print('|   MÉDIA DE IDADE         |       MEDIA DE ALTURA   |')
print('|   {:<8,.2f}               |              {:>8,.2f}   |' .format(media_idade,media_altura))
print ("-"*54)
```
